```python
from collections import defaultdict

s = input()
p = input()
p_dict = defaultdict(int)
for i in p.lower():
    p_dict[i] += 1
answer = []
intersection_dict = p_dict.copy()
current_dict = defaultdict(int)
for i in range(len(s)):
    if s[i] in p_dict:
        current_dict[s[i]] += 1
        if current_dict[s[i]] == p_dict[s[i]]:
            intersection_dict.pop(s[i])
        elif current_dict[s[i]] < p_dict[s[i]]:
            intersection_dict[s[i]] = p_dict[s[i]] - current_dict[s[i]]
    if i >= len(p):
        item = s[i - len(p)]
        if item in p_dict:
            if current_dict[item] != 0:
                current_dict[item] -= 1
            if current_dict[item] < p_dict[item]:
                intersection_dict[item] = p_dict[item] - current_dict[item]
    if len(intersection_dict) == 0:
        answer.append(i - len(p) + 1)
answer
```
